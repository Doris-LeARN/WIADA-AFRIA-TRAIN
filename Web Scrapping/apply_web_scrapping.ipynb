```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.jeuneafrique.com/"
```

```python
page_source = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(page_source.text,"html.parser")
```
